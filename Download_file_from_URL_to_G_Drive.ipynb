```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
import requests 
file_url="https://zenodo.org/record/3384388/files/-6_dB_fan.zip?download=1"
    
r = requests.get(file_url, stream = True) 

with open("/content/gdrive/MyDrive/challenge/fan22.zip", "wb") as file: 
    for block in r.iter_content(chunk_size = 1024): 
        if block: 
            file.write(block)
```
